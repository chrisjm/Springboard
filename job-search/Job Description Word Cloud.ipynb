```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
def build_wordcloud(filename):
    name = filename.replace(".txt","")
    text = ""
    
    with open(f"{filename}", encoding='utf-8') as f:
        text = ''.join(f.readlines())
    
    wc = WordCloud(width=800, height=600, background_color="white", max_words=50)
    wc.generate(text)
    wc.to_file(f"{name}_wordcloud.png")
    
    plt.imshow(wc, interpolation='bilinear')
    plt.title(name)
    plt.axis("off")
    plt.show()
```

```python
filename_list = ['data-analyst.txt', 'data-engineer.txt', 'data-scientist.txt']
for filename in filename_list:
    build_wordcloud(filename)
```
